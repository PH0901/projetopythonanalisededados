### Importação de Módulos e Arquivos

```python
import pandas as pd

funcionarios_df = pd.read_csv('CadastroFuncionarios.csv', sep = ';', decimal=',')
serviços_prestados_df = pd.read_excel('BaseServiçosPrestados.xlsx')
clientes_df = pd.read_csv('CadastroClientes.csv', sep = ';', decimal=',')

#retirar colunas Estado Civil e CArgo da tabela de funcionarios
funcionarios_df = funcionarios_df.drop(['Estado Civil', 'Cargo'], axis=1)

display(funcionarios_df)
display(serviços_prestados_df)
display(clientes_df)
```

### 1 -  Folha salarial

```python
funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']
print(f'Total da Folha Salarial Mensal é de R${funcionarios_df["Salario Total"].sum():,}')
display(funcionarios_df)
```

```python
#minha resolução (certa, mas maior)
salario_funcionarios = []

for salario_total_funcionario in funcionarios_df:
    salario_total_funcionario = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']
    salario_funcionarios.append(salario_total_funcionario)

salario_total_funcionarios = sum(salario_total_funcionario)
print(f'Total da Folha Salarial Mensal é de R${salario_total_funcionarios:,}')

funcionarios_df['Salario Total'] = salario_total_funcionario
display(funcionarios_df)
```

### 2 - Faturamento Empresa

```python
faturamentos_df = serviços_prestados_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['Valor Contrato Mensal', 'ID Cliente']], on='ID Cliente')
faturamentos_df['Faturamento Total'] = faturamentos_df['Tempo Total de Contrato (Meses)'] * faturamentos_df['Valor Contrato Mensal']
print(f'O faturamento total da empresa foi de R${faturamentos_df["Faturamento Total"].sum():,}')
display(faturamentos_df)
```

### 3 - % Funcionarios Fecharam Contrato

```python
funcionarios_totais = len(funcionarios_df['ID Funcionário'])
print(funcionarios_totais)

funcionarios_fecharam_contrato = len(serviços_prestados_df['ID Funcionário'].unique())
print(funcionarios_fecharam_contrato)

porc_funcionarios_fecharam_contrato = funcionarios_fecharam_contrato / funcionarios_totais
print(f'Percentual de funcionários que fecharam contrato: {porc_funcionarios_fecharam_contrato:.2%}')
```

### 4 - Contratos por Área

```python
contratos_area_df = serviços_prestados_df[['ID Funcionário']].merge(funcionarios_df[['ID Funcionário', 'Area']], on='ID Funcionário')
display(contratos_area_df)
contratos_area_qntd = contratos_area_df['Area'].value_counts()
print(contratos_area_qntd)
contratos_area_qntd.plot(kind='bar')
```

### 5 - Qntd Contratos por Área

```python
qntd_funcionarios_area = funcionarios_df[['Area']].value_counts()
print(qntd_funcionarios_area)
qntd_funcionarios_area.plot(kind='bar')
```

### 6 - Ticket Médio Mensal

```python
ticket_medio_mensal = clientes_df['Valor Contrato Mensal'].mean()
print(f'Ticket Médio Mensal: R${ticket_medio_mensal:,.2f}')
```
